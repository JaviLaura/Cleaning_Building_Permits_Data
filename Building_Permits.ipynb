<h3>Data Intuition</h3>

Data can be missing for one of two reasons:<br>
<ol>
<li>It doesn't exist</li>
<li>It wasn't recorded</li>
</ol>

For example, the "age of the oldest child" doesn't exist in a family that doesn't have any children. However, the "total number of living family members" will always exist, but may not be recorded. We can guess the values of unrecorded data, but it wouldn't make sense to guess for data that doesn't exist.

<h3>Clean the Data</h3>

Let's try cleaning the data in 3 different ways: dropping rows, dropping columns, and filling in data. <br>

```
#Delete rows
buildDFclean_1 = buildDF.dropna()
display(buildDFclean_1)
```

```
#Delete columns
buildDFclean_2 = buildDF.dropna(axis=1)
display(buildDFclean_2)
```

```
#Fill in missing values

buildDFclean_3 = buildDF.fillna(0)
display(buildDFclean_3)
```
